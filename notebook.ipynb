```javascript
const data = require('./output/au-sa.json');
const { entries, array2table } = require('./lib/helper');
```

## Some counts to start things off

```javascript
var totalNames = Object.keys(data.names).length;
var uniqueNames = entries(data.names).filter(name => name.value.count.total === 1);
var almostUniqueNames = entries(data.names).filter(name => name.value.count.total === 2);
var totalUniqueNames = uniqueNames.length;

$$html$$ = array2table([['Total names', totalNames],['Unique names', totalUniqueNames], ['Names used twice', almostUniqueNames.length]]);
```

```javascript
$$html$$ = array2table(entries(data.names.Simon.years)
    .sort((a, b) => +b.key - +a.key)
    .map(y => [y.key,y.value.prevalence.male]), ['Year', 'Prevalence']);
```

## Ranking of all names

```javascript
var nameRank = entries(data.names)
    .sort((a, b) => b.value.prevalence.total - a.value.prevalence.total)
    .map((v, i) => [i+1, v.key, v.value.count.total, v.value.prevalence.total]).slice(0, 10);
// nameRank;
$$html$$ = array2table(nameRank, ['Rank','Name','Count','Prevalence']);
```
